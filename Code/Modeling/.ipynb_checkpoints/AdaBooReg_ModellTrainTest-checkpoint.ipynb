#### Füge die Target- und Predictions-Spalten hinzu um mit dem Datensatz mit dem MLFlow-Run zu speichern

```python
target = adabooreg_run_1.df_pred[['date', 'calls_reg_act_diff']].copy()
df_only_features = adabooreg_run_1.df_only_features.copy()
df_features_target = pd.concat([df_only_features, target], axis=1)
print(df_features_target.head())
```

```python
df_features_target['predictions'] = adabooreg_run_1.model_adabooreg.adabooreg_pred_all
print(df_features_target.head())
```

#### X_test wird mit dem Modell in MLFlow gespeichert

```python
X_test = adabooreg_run_1.model_adabooreg.X_test
```

#### Definiere die Details für MLFlow

mlflow.set_tracking_uri("http://127.0.0.1:8080")

# Laufendes Experiment definieren
experiment_id = "AdaBooReg"
adabooreg_experiment = mlflow.set_experiment(experiment_id)

# run-name definieren
now = datetime.now()
now = now.strftime("%Y%m%d-%H%M")
run_name = f"AdaBooReg_{now}"

artifact_path = "bereitschaftsdienstplans"

```python
dataset = mlflow.data.from_pandas(df_features_target, targets='calls_reg_act_diff', predictions='predictions')
```

#### Log diesen Training-Run

# Initiate the MLflow run context
with mlflow.start_run(run_name=run_name) as run:

    # Datensatz protokollieren
    mlflow.log_input(dataset)
    
    # Parameter protokollieren
    mlflow.log_params(params)

    # Metriks protokollieren
    mlflow.log_metrics(metrics)

    # trainiertes Modell-Instanz protokollieren
    mlflow.sklearn.log_model(
        sk_model=fitted_model, 
        input_example=X_test,
        artifact_path=artifact_path
    )

mlflow.end_run()
